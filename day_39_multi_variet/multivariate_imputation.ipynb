KNNImputer 

The KNNImputer is a method for imputing missing values by using the k-nearest neighbors approach. Instead of using statistical measures like mean, median, or mode to estimate missing values, it leverages the values of the nearest neighbors to impute missing values. 

Here's how KNNImputer works:

1. **Identify Nearest Neighbors**: For each missing value, the algorithm identifies the k nearest neighbors based on the observed values of the other features in the dataset.

2. **Compute Imputed Value**: It then computes the imputed value for the missing value by taking the average (or weighted average) of the corresponding feature values of the nearest neighbors.

3. **Repeat for all Missing Values**: This process is repeated for all missing values in the dataset.

When to use KNNImputer:

1. **Continuous and Categorical Features**: KNNImputer can handle both continuous and categorical features, making it suitable for datasets with mixed data types.

2. **No Assumptions About Data Distribution**: KNNImputer makes no assumptions about the underlying distribution of the data, making it useful when data distribution is unknown or non-normal.

3. **Non-linear Relationships**: It can capture non-linear relationships between features, which may be missed by simple imputation methods.

Advantages of KNNImputer:

1. **Preserves Relationships**: KNNImputer preserves the relationships between features by using the values of the nearest neighbors, leading to more accurate imputations.

2. **Handles Mixed Data Types**: It can handle datasets with both numerical and categorical features without the need for separate imputation methods.

3. **No Distribution Assumptions**: KNNImputer does not assume any specific data distribution, making it robust to various types of data.

Disadvantages of KNNImputer:

1. **Computationally Intensive**: Computing nearest neighbors for each missing value can be computationally intensive, especially for large datasets or high-dimensional data.

2. **Sensitive to Outliers**: KNNImputer is sensitive to outliers, as they can significantly affect the nearest neighbor distances and consequently the imputed values.

3. **Parameter Selection**: The choice of the number of neighbors (k) can impact the imputation results, and selecting an appropriate value of k may require experimentation.

In summary, KNNImputer is a flexible imputation method suitable for datasets with mixed data types and no assumptions about data distribution. It can capture complex relationships between features but may be computationally intensive and sensitive to outliers. It's best used when other imputation methods are not appropriate or when preserving feature relationships is crucial.

How to calculate

Uniform Weight: 

When using uniform weights, all neighbors within the specified number of nearest neighbors (k) contribute equally to the imputed value. Regardless of their distance from the missing value, each neighbor's value is given the same weight in computing the imputed value. Mathematically, this means that all neighbors are weighted equally, and their contributions are averaged to compute the imputed value.


Distance Weight:
 
 When using distance weights, the contribution of each neighbor to the imputed value is weighted inversely proportional to its distance from the missing value. Closer neighbors have a higher influence on the imputed value, while farther neighbors have a lower influence. Mathematically, the weight assigned to each neighbor is determined by its distance from the missing value, with closer neighbors receiving higher weights and farther neighbors receiving lower weights. This ensures that nearby observations have a greater impact on the imputed value than distant ones.
