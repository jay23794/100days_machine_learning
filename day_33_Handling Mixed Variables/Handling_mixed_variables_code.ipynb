```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("titanic.csv")
```

```python
df.head()
```

```python
df['number'].unique()
```

```python
fig = df['number'].value_counts().plot.bar()
fig.set_title("passenger travelling with")
```

```python
# making number column data NAN where values are in Alphabetical format e.g A
df['number_numerical'] = pd.to_numeric(df['number'],errors='coerce')
```

```python
# making number column data NAN where values are in Number  format e.g 1,2, 3 
df['number_category']= df['number'].where(df['number'].str.isalpha())
```

```python
df.head()
```

```python
df['Cabin'].unique()
```

```python
df['Cabin'].notna().apply(lambda x :print(x))
```

```python
df['cabin_num']=df['Cabin'].str.extract('(\d+)')
df['cabin_cat']=df['Cabin'].str[0]
```

```python
df['ticket_num'] = df['Ticket'].apply(lambda x :x.split()[-1])
df['ticket_num'] = pd.to_numeric(df['ticket_num'],errors='coerce',downcast='integer')
df['ticket_cat']=df['ticket_cat'].apply(lambda x : np.nan if  x.isdigit() else x)
```
